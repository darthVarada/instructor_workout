```python
import pandas as pd

df_users = pd.read_parquet("users_silver_20251207.parquet")
df_heavy = pd.read_parquet("merged.parquet")
df_kaggle = pd.read_parquet("estudo_merge_userseHeavy.ipynb")

print("Users:", df_users.shape)
print("Heavy:", df_heavy.shape)
print("Kaggle:", df_kaggle.shape)
```

```python
users_ids = set(df_users["user_id"].dropna().unique())
heavy_users = set(df_heavy["user_id"].dropna().unique())

print("Users no cadastro:", len(users_ids))
print("Users no heavy:", len(heavy_users))

print("✅ Interseção:", len(users_ids & heavy_users))
print("❌ Heavy sem users:", len(heavy_users - users_ids))
```

```python
heavy_orfaos = df_heavy[~df_heavy["user_id"].isin(users_ids)]
heavy_orfaos[["user_id"]].drop_duplicates().head(10)
```

```python
import re
def normalize_text(x):
    if pd.isna(x):
        return x
    x = str(x).lower().strip()
    x = re.sub(r"\s+", " ", x)
    return x

df_heavy["exercise_norm"] = df_heavy["exercise_title"].apply(normalize_text)
df_kaggle["exercise_norm"] = df_kaggle["exercise_name"].apply(normalize_text)
```
